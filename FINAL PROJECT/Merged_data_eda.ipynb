```python
# How much missing data do we have per column?
cols_with_nulls = df.columns[df.isnull().any()]

idx = []
missing = []
missing_pct = []

for col in cols_with_nulls:
    idx.append(col)
    missing.append(len(df[col][df[col].isnull()]))
    missing_pct.append(100 * len(df[col][df[col].isnull()])/len(df[col]))

missing_df = pd.DataFrame(np.array([missing, missing_pct]).T, index=idx, columns=['num missing', 'pct missing'])
missing_df.head()
```

```python
missing_sorted = missing_df.sort_values(by=['pct missing'], ascending=False)
print(missing_sorted)
```
